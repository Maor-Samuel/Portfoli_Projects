```python
# Set style and context
sns.set_style("whitegrid")
sns.set_context("talk")  # bigger fonts

# Create the figure
plt.figure(figsize=(12, 6))
sns.histplot(df3['race_lenght'], bins=30, color='skyblue', edgecolor='black', kde=True)

# Titles and labels
plt.title('Distribution of Race Length', fontsize=20)
plt.xlabel('Race Length', fontsize=16)
plt.ylabel('Frequency', fontsize=16)

# Show plot
plt.tight_layout()
plt.show()
```

As we can see above :
This distribution might reflect trends in the running community—like how 50k is often the entry-level ultramarathon, while 50mi is more advanced, attracting fewer participants.

```python
# Set style and context
sns.set_style("whitegrid")
sns.set_context("talk")  

# Create the figure
plt.figure(figsize=(12, 6))

# Plot with hue
sns.histplot(
    data=df3,
    x='race_lenght',
    hue='athlete_gender',
    bins=30,
    kde=True,
    palette='Set2',  
    edgecolor='black',
    alpha=0.7  
)

# Titles and labels
plt.title('Race Length Distribution by Gender', fontsize=20)
plt.xlabel('Race Length (km or mi)', fontsize=16)
plt.ylabel('Frequency', fontsize=16)

# Final layout and show
plt.tight_layout()
plt.show()
```

As we can see above:
1. 50k is by far the most popular race length overall
Total participation = 13,000 (M) + 7,000 (F) = 20,000
That confirms it’s likely the most accessible or beginner-friendly ultramarathon.

2. Males participate more than females at both distances
At 50k: males outnumber females almost 2:1
At 50mi: males also outnumber females by a little more than 2:1
This suggests a gender gap in participation, especially in longer races.

3. Participation drops sharply from 50k to 50mi
Male: from 13,000 → 4,000 (around 70% drop)
Female: from 7,000 → 1,800 ( around 74% drop)
This means:

Longer distances deter both genders, but the drop-off is slightly more severe for females.
It could reflect differences in training access, race availability, cultural factors, or even preference.

My Thought's : 
Encourage female participation through outreach, support programs, or female-specific training clinics.
50k events could be used as a gateway to promote ultra-endurance running.
Investigate why people stop at 50k — is it physical limits, event availability, or interest?


```python
# Set the theme
sns.set_style("whitegrid")
sns.set_context("talk")

# Create displot
g = sns.displot(
    df3[df3['race_lenght'] == '50mi']['athlete_average_speed'],
    kind='hist',
    bins=30,
    kde=True,
    color='mediumseagreen',
    height=6,
    aspect=1.5
)

# Titles and axis labels
g.fig.suptitle('Distribution of Athlete Average Speed (50mi Races)', fontsize=20)
g.set_axis_labels('Average Speed (km/h)', 'Frequency')
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.9)  # To prevent title overlap

# Optional: Add vertical line for median or mean
mean_speed = df3[df3['race_lenght'] == '50mi']['athlete_average_speed'].mean()
plt.axvline(mean_speed, color='red', linestyle='--')
plt.text(mean_speed + 0.1, 10, f'Mean: {mean_speed:.2f} km/h', color='red')

plt.show()
```

As we can see above:
There are 800+ runners clustered around an average speed of ~6.2 km/h.
The mean speed is ~7.14 km/h, so the peak is just below average, but still in a comfortable range.
The distribution drops off sharply on both ends — especially beyond 10 km/h, where only ~100 runners are found.

This sharp drop-off suggests:
There’s a narrow “sweet spot” in endurance pacing.
6.2 km/h might represent a natural, sustainable pace that balances energy use, fatigue, and performance over ultra-long distances.

My Thought :
If I were to start running ultramarathons, I'd aim for around 6 km/h. Based on the data, that speed seems to be the most common and possibly the most sustainable for long distances. It's also a useful benchmark to track my fitness compared to the global running community.

```python

# Set style and context
sns.set_style("whitegrid")
sns.set_context("talk")

# Create figure
plt.figure(figsize=(14, 6))

# Violin plot
sns.violinplot(
    data=df3,
    x='race_lenght',
    y='athlete_average_speed',
    hue='athlete_gender',
    split=True,
    inner='quart',
    palette='Set2',
    linewidth=1.2
)

# Titles and labels
plt.title('Athlete Average Speed by Race Length and Gender', fontsize=20)
plt.xlabel('Race Length (km or mi)', fontsize=16)
plt.ylabel('Average Speed (km/h)', fontsize=16)

# Legend customization
plt.legend(title='Gender', title_fontsize=14, fontsize=12, loc='upper right')

# Improve layout
plt.tight_layout()
plt.show()
```

As we can see above:
at the 50km runs the average speed of males is around 7.3 and for the females is around 6.3 .

at the 50mi runs the average speed of males is around 6.5 and for the females around 6.2 .

1. Gender Gap Narrows in Longer Races
At 50k, the male-female speed gap is ~1.0 km/h
At 50mi, it reduces to just ~0.3 km/h
Interpretation: As distances increase, performance between genders becomes more similar. This might reflect factors like endurance equalizing over time, or different pacing strategies.

2. Everyone Slows Down in 50mi vs 50k
Males: from 7.3 → 6.5 km/h
Females: from 6.3 → 6.2 km/h
Interpretation: Males show a larger drop in average speed (11%), while females maintain a more consistent pace.

This could suggest that:
Males might push harder at shorter distances and fatigue more in ultras.
Females may pace more evenly, showing strong endurance sustainability.

My Thougth:
Investigate why females maintain speed better over longer distances — pacing strategy? training style?
Consider creating gender-specific pacing models or coaching strategies based on this endurance difference.
Race organizers or running apps could offer adaptive pacing plans personalized by gender and race length.

For new runners entering the world of long-distance or ultra running, starting at 6 km/h is both realistic and sustainable.
It aligns closely with thousands of experienced runners — making it a safe, proven starting pace.

```python
# Set the style
sns.set_style("whitegrid")
sns.set_context("talk")

# Create the lmplot
g = sns.lmplot(
    data=df3,
    x='athlete_age',
    y='athlete_average_speed',
    hue='athlete_gender',
    aspect=1.5,
    height=6,
    palette='Set2',
    scatter_kws={'alpha': 0.4, 's': 40},  # Fainter, smaller dots
    line_kws={'linewidth': 2}
)

# Add title and axis labels
g.fig.suptitle('Relationship Between Age and Average Speed by Gender', fontsize=20)
g.set_axis_labels('Athlete Age', 'Average Speed (km/h)', fontsize=16)

# Adjust layout
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.9)

# Show the plot
plt.show()
```

As we can see above:
 1. Peak Performance Ages: 20–40
Both males and females show their highest average running speeds between ages 20 and 40.

This suggests these are the prime athletic years, likely due to a combination of physical peak and training capacity.

 2. Performance Declines with Age — but Not Equally
As age increases, average speed gradually declines for both genders — a natural drop due to endurance, recovery time, and physical changes.

However, females maintain participation at older ages (60–80) more noticeably than males.

This might indicate greater long-term health and endurance sustainability in females, or stronger community/social engagement in older age groups.

 3. High-Speed Gap is Significant
In the 14–16 km/h range, you have:

~45 male runners

Only ~2 female runners

That's a massive difference, suggesting that elite-level speeds are more often achieved by male athletes — possibly due to physiological differences, or representation in competitive racing.

 4. Gender Speed Gap Narrows with Age
The average male speed starts around 9 km/h, and females around 8 km/h.

As age increases, the gender gap shrinks to just ~0.2 km/h in later years.

This supports the idea that females retain consistent pacing ability and may decline more gradually than males in terms of average speed.

 My Toughts:
If you're looking at long-term running ability, it's clear that while males may dominate in raw speed at younger ages, females show impressive consistency and longevity — especially in endurance-style events.
